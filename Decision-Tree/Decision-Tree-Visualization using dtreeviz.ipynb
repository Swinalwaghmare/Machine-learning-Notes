```python
from sklearn.datasets import *
from dtreeviz import *
from IPython.display import Image, display_svg, SVG
```

```python
classifier = tree.DecisionTreeClassifier(max_depth=2)
iris = load_iris()

X_train = iris.data[:,:2]
y_train = iris.target

classifier.fit(X_train, y_train)
```

```python
# Sklearn Tree Method
from sklearn.tree import plot_tree
plot_tree(classifier)
```

### dtreeviz - Visualization (Classfication)

```python
viz = dtreeviz.model(classifier,X_train, y_train,target_name='price',
               feature_names=iris.feature_names,class_names=['setosa', 'versicolor', 'virginica'])
viz
```

```python
from dtreeviz import decision_boundaries
```

```python
decision_boundaries(classifier, X_train, y_train, ntiles=50, binary_threshold=.5,
                   markers=['X','s'], feature_names=iris.feature_names)
```
