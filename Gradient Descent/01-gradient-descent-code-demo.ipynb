```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.show()
```

```python
# Lets apply OLS
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(X,lr.predict(X),color='red')
plt.show()
```

```python
# Lets apply gradient descent assuming slope in constant m = 78.35
# and let's assume the starting value of intercept b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(X,lr.predict(X),color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0 

loss_slope = -2 * np.sum(y - m*X.ravel()-b)
loss_slope
```

```python
# Lets take learning rate = 0.1
learning_rate = 0.1

step_size = loss_slope * learning_rate
step_size
```

```python
# caculating the new intercept

b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(X,lr.predict(X),color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X, y_pred1, color='black', label='{}'.format(b))
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)

loss_slope
```

```python
step_size = loss_slope * learning_rate
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(X,lr.predict(X),color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X, y_pred2, color='black', label='{}'.format(b))
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * learning_rate
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(X,lr.predict(X),color='red', label='OLS')
plt.plot(X, y_pred, color='blue', label='b=0')
plt.plot(X, y_pred3, color='black', label='{}'.format(b))
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    step_size = lr*loss_slope
    b = b - step_size

    y_pred = m * X + b

    plt.plot(X, y_pred)

plt.scatter(X,y)
```
